```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import time

def loadData():
    with np.load('notMNIST.npz') as data :
        Data, Target = data ['images'], data['labels']
        posClass = 2
        negClass = 9
        dataIndx = (Target==posClass) + (Target==negClass)
        Data = Data[dataIndx]/255.
        Target = Target[dataIndx].reshape(-1, 1)
        Target[Target==posClass] = 1
        Target[Target==negClass] = 0
        np.random.seed(421)
        randIndx = np.arange(len(Data))
        np.random.shuffle(randIndx)
        Data, Target = Data[randIndx], Target[randIndx]
        trainData, trainTarget = Data[:3500], Target[:3500]
        validData, validTarget = Data[3500:3600], Target[3500:3600]
        testData, testTarget = Data[3600:], Target[3600:]
    return trainData, validData, testData, trainTarget, validTarget, testTarget

```

```python
trainData, validData, testData, trainTarget, validTarget, testTarget = loadData()
print(np.shape(trainData))
print(trainTarget[1], trainTarget[2]) # C->1
# plt.imshow(trainData[1])
plt.imshow(trainData[2])
```

```python
def MSE(W, b, x, y, reg):
    loss = 0 
    for i in range(0, len(x)):
        temp = np.reshape(x[i], (len(x[i])**2, 1))
        loss +=  np.linalg.norm((np.matmul(W.transpose(), temp) + b - y[i]))**2
    loss *= 1/(2*len(x)) 
    loss += reg/2*np.linalg.norm(W)**2
    return loss

W = np.ones((784,1))
MSE(W, 1, trainData, trainTarget, 1)

# x = trainData[0]
# y = np.reshape(x, (784, 1))
# z = np.dot(W.transpose(), y)
# x->trainData, y->trainTarget
```

```python
def gradMSE(W, b, x, y, reg):
#     weightsGradient = np.zeros((len(x[0])**2+1,1))
    modified_W = np.insert(W, 0, b, axis=0)
    
    new_x = []
    for i in range(0, len(x)):
        new_x_i = np.reshape(x[i], (len(x[i])**2, 1))
        new_x_i = np.insert(new_x_i, 0, 1, axis=0)
        new_x.append(new_x_i)
    new_x = np.array(new_x)
    new_x_2 = np.reshape(new_x,(3500, 785))

    weightsGradient =  -2*np.matmul(new_x_2.transpose(),y) 
    weightsGradient += 2*new_x_2.transpose().dot(new_x_2).dot(modified_W)

    weightsGradient *= 1/(2*len(x)) 
    weightsGradient += reg * modified_W
    
    biasGradient = weightsGradient[0]
    weightsGradient = weightsGradient[1:]
    return biasGradient, weightsGradient

W = np.zeros((784,1))
gradMSE(W, 1, trainData, trainTarget, 0)
```

```python
def grad_descent(W, b, trainingData, trainingLabels, alpha, iterations, reg, EPS):
#     Your function must return the trained weights and biases. You may also wish to print and
# store the training, validation and test losses/accuracies in this function for plotting.
    old_error = MSE(W, b, trainingData, trainingLabels, reg)
    loss = []
    loss.append(old_error)
    start_time = time.time()
    for i in range(0, iterations):
        bias_gradient, weights_gradient = gradMSE(W, b, trainingData, trainingLabels, reg)
        W = W - alpha * weights_gradient
        b = b - alpha * bias_gradient
        new_error = MSE(W, b, trainingData, trainingLabels, reg)
        loss.append(new_error)
        if abs(old_error - new_error) < EPS:
            print(old_error, new_error)
            break
        old_error = new_error
    end_time = time.time()
    elapsed_time = end_time - start_time
    print("Total time elapsed: {:.2f} seconds".format(elapsed_time))
    return loss

W = np.zeros((784,1))
b = 1
grad_descent(W, b, trainData, trainTarget, 0.005, 20, 0, 10**-7)
```

```python
def get_acc(W, b, data, labels, reg):
    x = data
    y = labels
    num_correct = 0
    for i in range(0, len(x)):
        temp = np.reshape(x[i], (len(x[i])**2, 1))
        prediction = np.matmul(W.transpose(), temp) + b
        actual = y[i]
        
        if actual == 1 and prediction - actual > -0.5 or actual == 0 and prediction - actual < 0.5:
            num_correct += 1
#         loss +=  np.linalg.norm((np.matmul(W.transpose(), temp) + b - y[i]))**2
#     loss *= 1/(2*len(x)) 
#     loss += reg/2*np.linalg.norm(W)**2
    return num_correct / len(x)
    
```

```python
def grad_descent2(W, b, trainingData, trainingLabels, alpha, iterations, reg, EPS, validationData, validationLabels, testData=None, testLabels=None):
#     Your function must return the trained weights and biases. You may also wish to print and
# store the training, validation and test losses/accuracies in this function for plotting.
    old_loss_train = MSE(W, b, trainingData, trainingLabels, reg)
    old_loss_val = MSE(W, b, validationData, validationLabels, reg)
    old_loss_test = MSE(W, b, testData, testLabels, reg)
    
    old_acc_train = get_acc(W, b, trainingData, trainingLabels, reg)
    old_acc_val = get_acc(W, b, validationData, validationLabels, reg)
    old_acc_test = get_acc(W, b, testData, testLabels, reg)
    
    
    loss_train = [old_loss_train]
    loss_valid = [old_loss_val]
    loss_test = [old_loss_test]
    
    acc_train = [old_acc_train]
    acc_valid = [old_acc_val]
    acc_test = [old_acc_test]
    
    
    start_time = time.time()
    
    for i in range(0, iterations):
        bias_gradient, weights_gradient = gradMSE(W, b, trainingData, trainingLabels, reg)
        W = W - alpha * weights_gradient
        b = b - alpha * bias_gradient
        new_loss_train = MSE(W, b, trainingData, trainingLabels, reg)
        new_loss_valid = MSE(W, b, validationData, validationLabels, reg)
        new_loss_test = MSE(W, b, testData, testLabels, reg)
        
        new_acc_train = get_acc(W, b, trainingData, trainingLabels, reg)
        new_acc_valid = get_acc(W, b, validationData, validationLabels, reg)
        new_acc_test = get_acc(W, b, testData, testLabels, reg)
        
        loss_train.append(new_loss_train)
        loss_valid.append(new_loss_valid)
        loss_test.append(new_loss_test)
        
        acc_train.append(new_acc_train)
        acc_valid.append(new_acc_valid)
        acc_test.append(new_acc_test)
        
        if abs(old_loss_train - new_loss_train) < EPS:
            print(old_loss_train, new_loss_train)
            break
        old_loss_train = new_loss_train
        
    end_time = time.time()
    elapsed_time = end_time - start_time
    print("Total time elapsed: {:.2f} seconds".format(elapsed_time))
    
    return loss_train, acc_train, loss_valid, acc_valid, loss_test, acc_test


W = np.zeros((784,1))
b = 1
epochs = 5000
reg = 0
learning_rate_big = 0.005
learning_rate_medium = 0.001
learning_rate_small = 0.0001
training_loss_big, acc_train_big, validation_loss_big, acc_valid_big, test_loss_big, acc_test_big = grad_descent2(W, b, trainData, trainTarget, learning_rate_big, epochs, 0, 10**-7, validData, validTarget, testData, testTarget)
W = np.zeros((784,1))
training_loss_medium, acc_train_medium, validation_loss_medium, acc_valid_medium, test_loss_medium, acc_test_medium = grad_descent2(W, b, trainData, trainTarget, learning_rate_medium, epochs, 0, 10**-7, validData, validTarget, testData, testTarget)
W = np.zeros((784,1))
training_loss_small, acc_train_small, validation_loss_small, acc_valid_small, test_loss_small, acc_test_small = grad_descent2(W, b, trainData, trainTarget, learning_rate_small, epochs, 0, 10**-7, validData, validTarget, testData, testTarget)


```

```python
def normalEquation(W, b, x, y, reg):
    start_time = time.time()
    
    new_x = []
    for i in range(0, len(x)):
        new_x_i = np.reshape(x[i], (len(x[i])**2, 1))
        new_x_i = np.insert(new_x_i, 0, 1, axis=0)
        new_x.append(new_x_i)
    new_x = np.array(new_x)
    new_x = np.reshape(new_x,(3500, 785))
    
    first_term = np.linalg.inv(np.matmul(new_x.transpose(), new_x))
    second_term = np.matmul(new_x.transpose(), y)
    optimal_weights = np.matmul(first_term, second_term)
    
    old_acc_train = get_acc(optimal_weights[1:], optimal_weights[0], x, y, None)
    print(old_acc_train)
    
    end_time = time.time()
    elapsed_time = end_time - start_time
    print("Total time elapsed: {:.2f} seconds".format(elapsed_time))
    print(MSE(optimal_weights[1:], optimal_weights[0], x, y, reg))

W = np.zeros((784,1))
b = 1
normalEquation(W, b, trainData, trainTarget, 0)
```

```python
def crossEntropyLoss(W, b, x, y, reg):
    loss = 0 
    for i in range(0, len(x)):
        new_x = np.reshape(x[i], (len(x[i])**2, 1))
        y_hat = 1/(1+np.exp(-(np.matmul(W.transpose(), new_x) + b)))
        loss -=  (y[i]*np.log(y_hat) + (1-y[i])*np.log(1-y_hat))
    loss *= 1/len(x) 
    loss += reg/2*np.linalg.norm(W)**2
    return np.asscalar(loss)

W = np.zeros((784,1))
b = 1
crossEntropyLoss(W, b, trainData, trainTarget, 0)
```

```python
def grad_descent(W, b, trainingData, trainingLabels, alpha, iterations, reg, EPS, lossType="None"):
#     Your function must return the trained weights and biases. You may also wish to print and
# store the training, validation and test losses/accuracies in this function for plotting.
    acc = [get_acc(W, b, trainingData, trainingLabels, reg)]
    
    if lossType=="MSE":
        old_loss = MSE(W, b, trainingData, trainingLabels, reg)
        loss = [old_loss]
        
        for i in range(0, iterations):
            bias_gradient, weights_gradient = gradMSE(W, b, trainingData, trainingLabels, reg)
            W = W - alpha * weights_gradient
            b = b - alpha * bias_gradient
            
            new_loss = MSE(W, b, trainingData, trainingLabels, reg)
            new_acc = get_acc(W, b, trainingData, trainingLabels, reg)
            
            acc.append(new_acc)
            loss.append(new_loss)
            if abs(old_loss - new_loss) < EPS:
                print(old_loss, new_loss)
                break
            old_loss = new_loss
            
        return loss, acc
    
    elif lossType=="CE":
        old_loss = crossEntropyLoss(W, b, trainingData, trainingLabels, reg)
        loss = [old_loss]
        
        for i in range(0, iterations):
            bias_gradient, weights_gradient = gradCE(W, b, trainingData, trainingLabels, reg)
            W = W - alpha * weights_gradient
            b = b - alpha * bias_gradient
            
            new_loss = crossEntropyLoss(W, b, trainingData, trainingLabels, reg)
            new_acc = get_acc(W, b, trainingData, trainingLabels, reg)
            
            acc.append(new_acc)
            loss.append(new_loss)
            if abs(old_loss - new_loss) < EPS:
                print(old_loss, new_loss)
                break
            old_loss = new_loss
            
        return loss, acc
        

W = np.zeros((784,1))
b = 1
epochs = 5000
reg = 0.1
crossEntropyLoss, acc_train = grad_descent(W, b, trainData, trainTarget, 0.005, epochs, reg, 10**-7, "CE")
```

```python
def buildGraph(lossType=None):
    #Initialize weight and bias tensors
    tf.set_random_seed(421)
#     bias = tf.Variable(tf.ones(shape=(1)), name="bias")
    bias = tf.Variable(tf.random_normal([1]), name='bias')
    x_data = tf.placeholder(tf.float32, name="x_data")
    y_labels = tf.placeholder(tf.float32, name="y_labels")
    learning_rate = tf.placeholder(tf.float32, name="learning_rate")
    reg = tf.placeholder(tf.float32, name="reg")
#     reg = 0
    weights = tf.Variable(tf.truncated_normal(shape=[784,1], stddev=0.5), name="weights")
    optimizer = tf.train.GradientDescentOptimizer(learning_rate)

    predictions = tf.add(tf.matmul(x_data, weights), bias)
    if lossType == "MSE":
        loss_tensor = tf.divide(tf.losses.mean_squared_error(y_labels, predictions), 1)
        loss_tensor += tf.multiply(reg / 2, tf.reduce_sum(tf.square(weights)))
    elif lossType == "CE":
        loss_tensor = tf.divide(tf.reduce_mean(tf.losses.sigmoid_cross_entropy(y_labels, predictions)), 1)
        loss_tensor += reg/2 *tf.linalg.norm(weights)**2    
    return weights, bias, predictions, y_labels, loss_tensor, optimizer, reg, x_data, learning_rate
```
